```
import os
import re
from collections import defaultdict, Counter
```

```
import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
nltk.download('stopwords')
mystopwords = stopwords.words('russian')
token = RegexpTokenizer('\w+')
```

# Считаем TF для контрастной коллекции

```
contrast_collection = []
```

```
for docs in os.walk('drive/My Drive/контрастный_корпус'):
  for doc in docs[2]:
    with open('drive/My Drive/контрастный_корпус/'+doc, 'r', encoding='utf-8') as f:
      text = f.read()
      contrast_collection.append(text)
```

```
contrast_collection
```

```
def tokenize(text):
    return token.tokenize(text)

def clear(text):
  reg = re.compile('[^А-яа-я- ]')
  text = re.sub(reg, '', text)
  return text
    
def normalize(text):
    words = [morph.parse(word)[0].normal_form for word in tokenize(clear(text)) if word]
    return words
```

```
contrast_termdict = {}
for idx, text in enumerate(contrast_collection, start=1):
        text = [word for word in normalize(text) if word not in mystopwords]
        contrast_termdict[idx] = text
```

```
for i in list(contrast_termdict)[0:3]:
    print ("key: {}, value: {} ".format(i,  contrast_termdict[i]))
```

```
inv_idx_contrast = {}
for k,value in contrast_termdict.items():
    for v in value:
        if v not in inv_idx_contrast:
            inv_idx_contrast[v] = []
            inv_idx_contrast[v].append(k)
        else:
            list_of_values = list(inv_idx_contrast.get(v))
            list_of_values.append(k)
            inv_idx_contrast[v] = list_of_values
```

```
for i in list(inv_idx_contrast)[0:3]:
    print ("key: {}, value: {} ".format(i,  inv_idx_contrast[i]))
```

```
length_contrast = 0
for sent in contrast_collection:
  length_contrast += len(sent)
```

```
length_contrast
```

```
c=Counter(inv_idx_contrast)
#создаем словарь {key=слово : value=сколько раз встречается}
freq_dict_contrast = {}
for i in c.items():
    freq_dict_contrast[i[0]] = len(i[1])/length_contrast
```

```
for i in list(freq_dict_contrast)[0:3]:
    print ("key: {}, value: {} ".format(i,  freq_dict_contrast[i]))
```

# Считаем TF для исследуемой коллекции

```
import pandas as pd
```

```
test_collection = pd.read_excel('drive/My Drive/Test_автобрея.xlsx')
```

```
test_collection
```

```
test_termdict = {}
for idx, text in enumerate(test_collection['text'], start=1):
        text = [word for word in normalize(text) if word not in mystopwords]
        test_termdict[idx] = text
```

```
for i in list(test_termdict)[0:3]:
    print ("key: {}, value: {} ".format(i,  test_termdict[i]))
```

```
inv_idx_test = {}
for k,value in test_termdict.items():
    for v in value:
        if v not in inv_idx_test:
            inv_idx_test[v] = []
            inv_idx_test[v].append(k)
        else:
            list_of_values = list(inv_idx_test.get(v))
            list_of_values.append(k)
            inv_idx_test[v] = list_of_values
```

```
for i in list(inv_idx_test)[0:3]:
    print ("key: {}, value: {} ".format(i,  inv_idx_test[i]))
```

```
length_test = 0
for sent in test_collection['text']:
  length_test += len(sent)
```

```
length_test
```

```
c=Counter(inv_idx_test)
#создаем словарь {key=слово : value=сколько раз встречается}
freq_dict_test = {}
for i in c.items():
    freq_dict_test[i[0]] = len(i[1])/length_test
```

```
for i in list(freq_dict_test)[0:3]:
    print ("key: {}, value: {} ".format(i,  freq_dict_test[i]))
```

# Находим слова, входящие в обе коллекции

Частотные слова для исследуемой коллекции:

```
from statistics import mean
```

```
mean_test = mean(freq_dict_test[k] for k in freq_dict_test) 
```

```
mean_test
```

```
most_freq_test = {}
for word in freq_dict_test:
  if freq_dict_test.get(word) > mean_test:
    most_freq_test[word] = freq_dict_test.get(word)
```

```
for i in list(most_freq_test)[0:3]:
    print ("key: {}, value: {} ".format(i,  most_freq_test[i]))
```

```
test_contrast = set(most_freq_test).intersection(set(freq_dict_contrast))
```

```
test_contrast
```

# Считаем странность

**Формула**: (FRL/FRC)/(FRLC/FRCC)

*   FRL - частотность леммы в исследуемой коллекции
*   FRC - число словоупотреблений в исследуемой коллекции
*   FRLC - частотность леммы в контрастной коллекции
*   FRCC - число словоупотреблений в контрастной коллекции

```
words_strangeness = {}
for word in test_contrast:
  strangeness = freq_dict_test.get(word)/freq_dict_contrast.get(word)
  words_strangeness[word] = strangeness
```

```
for i in list(words_strangeness)[0:3]:
    print ("key: {}, value: {} ".format(i,  words_strangeness[i]))
```

```
mean_strange = mean(words_strangeness[k] for k in words_strangeness) 
```

```
mean_strange 
```

```
stranger_words = {}
for key, value in words_strangeness.items():
  if value > mean_strange:
    stranger_words[key] = value
```

```
for i in list(stranger_words)[0:3]:
    print ("key: {}, value: {} ".format(i,  stranger_words[i]))
```

# * Записываем в csv файл

```
import csv
from itertools import zip_longest
```

```
keys = []
values = []
for key, value in stranger_words.items():
  keys.append(key)
  values.append(value)
```

```
d = [keys, values]
export_data = zip_longest(*d, fillvalue = '')
with open('Stranger_Words.csv', 'w', encoding="utf-8", newline='') as f:
      wr = csv.writer(f)
      wr.writerow(("Word", "Strangeness"))
      wr.writerows(export_data)
```
